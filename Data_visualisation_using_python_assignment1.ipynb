```
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x=np.arange(0,10)
y=x*x
```

```
plt.title("line plot",)
plt.xlabel("x-axis",color='blue')
plt.ylabel("y-axis",color='red')
```

```
plt.plot(x,y,'r*',linestyle='dashed',linewidth=3,markersize=5)
```
